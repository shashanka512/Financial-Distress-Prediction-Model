```python
import pandas as pd
import numpy as np
```

```python
file ='Best_Model.xlsx'
```

```python
DFpivot = pd.read_excel(file, sheet_name='DFpivot', index_col=0, header=0)
DFpivot = pd.DataFrame(DFpivot)


DFpivot['Total'] = DFpivot['Active'] + DFpivot['Failure']
print(DFpivot)
```

```python
file1 = 'Y_pred6pviot1.xlsx'
Y_pred = pd.read_excel(file1, sheet_name='Sheet1', index_col=0, header=0)
Y_pred['Total'] = DFpivot['Total']
Y_pred['Failure'] = Y_pred['Total'] - Y_pred['Active']
print(Y_pred)

```

```python
Y_pred_yearly = pd.DataFrame()
Y_pred_yearly['Failure'] = Y_pred.resample('YE')['Failure'].mean()
Y_pred_yearly['Active'] = Y_pred.resample('YE')['Active'].mean()
Y_pred_yearly
```

```python

print(DFpivot)
```

```python

DF_pivot_yearly = pd.DataFrame()
```

```python
DF_pivot_yearly['Active'] = DFpivot.resample('Y')['Active'].mean()
DF_pivot_yearly['Failed'] = DFpivot.resample('Y')['Failure'].sum()
print(DF_pivot_yearly)
```

```python
DF_povot_yearly = DFpivot.resample('Y').agg({'Active':'mean', 'Failure':'sum'})
print(DF_povot_yearly)
```

```python
Failure_rate = pd.DataFrame()
Failure_rate['Actual'] = DFpivot['Failure'] / (DFpivot['Failure'] + DFpivot['Active'])
Failure_rate['Predicted'] = Y_pred['Failure'] / (Y_pred['Failure'] + Y_pred['Active'])
Failure_rate
```

```python
Failure_rate_yearly = pd.DataFrame()
Failure_rate_yearly['Actual'] = Failure_rate.resample('Y')['Actual'].mean()
Failure_rate_yearly['Predicted'] = Failure_rate.resample('Y')['Predicted'].mean()
Failure_rate_yearly


```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10, 6))

plt.plot(Failure_rate_yearly['Actual'], label='Actual', scalex=True, scaley=True)
plt.plot(Failure_rate_yearly['Predicted'], label='Predicted', scalex=True, scaley=True)
plt.title('Failure Rate (Yearly)')
plt.xlabel('Year')
plt.ylabel('Failure Rate')


plt.legend()
plt.show()

```
