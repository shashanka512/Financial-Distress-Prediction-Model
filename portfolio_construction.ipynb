```python
import pandas as pd
import numpy as np
```

```python
file = 'All.xlsx'
```

```python
ER = pd.read_excel(file, sheet_name='ER')
ER = pd.melt(ER, id_vars=['company'], var_name='Date', value_name='ER')
ER['Date'] = pd.to_datetime(ER['Date'])
ER = ER.set_index(['Date', 'company'])
ER = ER.sort_index()
 
print(ER.head())
print(ER.describe())
```

```python
MV = pd.read_excel(file, sheet_name='M')
MV = pd.melt(MV, id_vars=['company'], var_name='Date', value_name='MV')
MV['Date'] = pd.to_datetime(MV['Date'])
MV = MV.set_index(['Date', 'company'])
MV = MV.sort_index()

print(MV.head())
print(MV.describe())
```

```python
MB = pd.read_excel(file, sheet_name='MB')
MB = pd.melt(MB, id_vars=['company'], var_name='Date', value_name='MB')
MB['Date'] = pd.to_datetime(MB['Date'])
MB = MB.set_index(['Date', 'company'])
MB = MB.sort_index()

print(MB.head())
print(MB.describe())
```

```python
RSIZE = pd.read_excel(file, sheet_name='RSIZE')
RSIZE = pd.melt(RSIZE, id_vars=['company'], var_name='Date', value_name='RSIZE')
RSIZE['Date'] = pd.to_datetime(RSIZE['Date'])
RSIZE = RSIZE.set_index(['Date', 'company'])
RSIZE = RSIZE.sort_index()

print(RSIZE.head())
print(RSIZE.describe())
```

```python
Prob = pd.read_excel(file, sheet_name='Prob', header=0, index_col=[0,1])
print(Prob.head())
print(Prob.describe())

```

```python
file_path = 'Factors.xlsx'
```

```python
factor = pd.read_excel(file_path, sheet_name='Sheet1')
factor['Date'] = pd.to_datetime(factor['Date'])
factor = factor.set_index(['Date'])
factor = factor.sort_index()
factor = factor.fillna(method='ffill')

print(factor.head())
print(factor.describe())
print(factor.shape)

```

```python
RF = pd.read_excel(file, sheet_name='RF')
RF = pd.melt(RF, id_vars=['company'], var_name='Date', value_name='RF')
RF['Date'] = pd.to_datetime(RF['Date'])
RF = RF.drop('company', axis=1)
RF = RF.set_index(['Date'])
RF = RF.sort_index()

print(RF.head())
print(RF.describe())


```

```python
RM = pd.read_excel(file, sheet_name='RM')
RM = pd.melt(RM, id_vars=['company'], var_name='Date', value_name='RM')
RM['Date'] = pd.to_datetime(RM['Date'])
RM = RM.drop(columns=['company'])
RM = RM.set_index(['Date'])
RM = RM.sort_index()

print(RM.head())

print(RM.describe())
```

```python
DF = pd.merge(Prob, ER, how='inner', left_index=True, right_index=True)
DF = pd.merge(DF, MB, how='inner', left_index=True, right_index=True)
DF = pd.merge(DF, RSIZE, how='inner', left_index=True, right_index=True)
DF = pd.merge(DF, MV, how='inner', left_index=True, right_index=True)
print(DF.head())
print(DF.describe())

```
